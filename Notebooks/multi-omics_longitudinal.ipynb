```python
import biom
import pandas as pd
```

```python
# Function to load BIOM table, sort rows by sum
def biom_to_df(biom_path):

    # Load BIOM table and convert to a DataFrame
    table = biom.load_table(biom_path)

    df = pd.DataFrame(table.matrix_data.toarray(),
                      index=table.ids(axis='observation'),
                      columns=table.ids(axis='sample'))
    
    
    # Sort rows by row sum in descending order
    df['row_sum'] = df.sum(axis=1)
    df = df.sort_values(by='row_sum', ascending=False)
    
    # Drop the 'row_sum' column before proceeding
    df = df.drop(columns=['row_sum'])

    # Transpose df
    df = df.T
    
    return df
```

```python
metaG_mags_path = '../Data/metaG/Tables/metaG_mags_species.biom'
metaG_mags_df = biom_to_df(metaG_mags_path)
metaG_mags_df
```

```python
wgs_md = pd.read_csv('../Metadata/wgs_simplied_md.tsv', sep='\t')
PP_304_data = wgs_md[wgs_md['subject_ID'] == 'PP_304']
PP_304_data = PP_304_data.sort_values(by='local_lesion_severity')
PP_304_data
```

```python
PP_304_samples = PP_304_data['sample_name'].to_list()
PP_304_samples
```

```python
subset_metaG = metaG_mags_df[metaG_mags_df.index.isin(PP_304_samples)]

# Create a mapping from sample names to their order position
order_mapping = {sample: i for i, sample in enumerate(PP_304_samples)}

# Assign an order column and sort by it, then drop the helper column
ordered_metaG = (subset_metaG
                 .assign(order=subset_metaG.index.map(order_mapping))
                 .sort_values('order')
                 .drop('order', axis=1))
ordered_metaG
```
