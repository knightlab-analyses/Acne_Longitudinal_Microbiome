```python
# Replace with the actual column name used for grouping (e.g., 'skin_status')
group_col = 'cohort'  

# Replace with your actual subject ID column
if 'subject_ID' in metadata.columns:
    id_col = 'subject_ID'
elif 'host_subject_ID' in metadata.columns:
    id_col = 'host_subject_ID'
else:
    raise ValueError("No subject ID column found in metadata.")

# Count unique individuals in each group
group_counts = metadata.groupby(group_col)[id_col].nunique()

# Print the counts
print("Number of unique individuals per group:")
for group, count in group_counts.items():
    print(f"{group}: {count} individuals")
```

```python
# Function to load BIOM table, sort rows by sum
def biom_to_df(biom_path):

    # Load BIOM table and convert to a DataFrame
    table = biom.load_table(biom_path)

    df = pd.DataFrame(table.matrix_data.toarray(),
                      index=table.ids(axis='observation'),
                      columns=table.ids(axis='sample'))
    
    return df
```
