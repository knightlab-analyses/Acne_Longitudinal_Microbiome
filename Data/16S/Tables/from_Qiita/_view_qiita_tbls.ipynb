```python
# Import Python packages
import pandas as pd
import numpy as np
import biom
from biom import Table
from biom.util import biom_open
from biom import load_table
import h5py
import qiime2 as q2
from qiime2 import Artifact
```

```python
def load_biom_table(biom_path):
    # Load BIOM table and convert to a DataFrame
    table = biom.load_table(biom_path)
    df = pd.DataFrame(table.matrix_data.toarray(),
                      index=table.ids(axis='observation'),
                      columns=table.ids(axis='sample'))

    
    return df
```

```python
def df_to_biom(df, out_biom_path):
    table = Table(
        df.values,
        observation_ids=df.index.astype(str).tolist(),
        sample_ids=df.columns.astype(str).tolist()
    )

    with h5py.File(out_biom_path, "w") as h5f:
        table.to_hdf5(
            h5f,
            generated_by="column intersection filter"
        )
```

```python
# Import direct tables from Qiita
path_16S_V1V3 = "179426_feature-table_16S_V1V3_rare-11054.biom"
path_16S_V4 = "174951_feature-table_16S_V4_rare-3769.biom"

df_16S_V1V3 = load_biom_table(path_16S_V1V3)
df_16S_V4 = load_biom_table(path_16S_V4)
```

```python
# Remove any samples that are blanks
df_16S_V1V3 = df_16S_V1V3.loc[
    :, ~df_16S_V1V3.columns.str.contains("BLANK", case=False, na=False)
]

df_16S_V4 = df_16S_V4.loc[
    :, ~df_16S_V4.columns.str.contains("BLANK", case=False, na=False)
]
```

```python
# Remove the Qiita artifact id prefix so sample ids are aligned between V1-V3 and V4
df_16S_V1V3.columns = df_16S_V1V3.columns.str.replace(
    r"^173980\.14901\.", "", regex=True
)

df_16S_V4.columns = df_16S_V4.columns.str.replace(
    r"^173620\.14901\.", "", regex=True
)
```

```python
# Save them as new bioms for downstream use
df_to_biom(df_16S_V1V3, "179426_feature-table_16S_V1V3_rare-11054_sampleIDfixed.biom")
df_to_biom(df_16S_V4,   "174951_feature-table_16S_V4_rare-3769_sampleIDfixed.biom")
```
